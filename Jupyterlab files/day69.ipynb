```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
array1=np.random.randint(1,100,10)
array1
```

```python
print(np.sum(array1))
print(np.mean(array1))
print(np.median(array1))
print(np.quantile(array1,0.5))
```

```python
plt.boxplot(array1, showmeans = True)
plt.ylim([-20,120])
plt.show()
```
